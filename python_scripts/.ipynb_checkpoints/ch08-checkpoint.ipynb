```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

## Dissipation-driven melting and compaction

The thickness of the layer of ice is the solution of equations

\begin{align}
\label{eq:cmp-diss-mech-odes-nondim-a}
\dot{\layerthick} &= -\cmpdisspar\layerthick\por,\\
\label{eq:cmp-diss-mech-odes-nondim-b}
\dot{\por} &= \e^{-\lambda\por} - \cmpdisspar\omp\por,
\end{align}

where $\cmpdisspar$, the *compaction-dissipation number*, is given by 

\begin{equation*}
  \cmpdisspar\equiv\frac{\density\sol\latent\layerstress\layerthick_0^2}
  {\eta_0^2\layershear_0^2},
\end{equation*}

The Python code below implements the numerical solutions of equations \eqref{eq:cmp-diss-mech-odes-nondim-a} and \eqref{eq:cmp-diss-mech-odes-nondim-b}.

```python
def thick(y, t, pi, lam):  # y = [H, phi]
    H = y[0]
    phi = y[1]
    return [-pi * H * phi,
            np.exp(-lam*phi) - pi * (1. - phi) * phi]
```

The figure below plots the layer thickness $\layerthick$ relative to initial versus time for three values of $\cmpdisspar$.

```python
f, ax = plt.subplots()
zoom = 2.0
f.set_size_inches(5.4 * zoom, 3.1 * zoom)
f.set_facecolor('w')

y0 = [1., 0.0]  # initial condition: H=1, phi=0.0
lambda_ = 27.0
Pi = np.asarray([0.1, 1., 10.])
tmax = np.asarray([60., 10., 2.6])
linesty = ['-', '--', '-.']
for pi, tm, ls in zip(Pi, tmax, linesty):
    t = np.arange(0.0, tm, 0.05)
    S = odeint(thick, y0, t, args=(pi, lambda_), rtol=1e-8)
    plt.semilogx(t, S[:, 0], 'k', linestyle=ls, linewidth=2, label=f'$\Pi={str(pi)}$')

plt.xlabel(r'$t/[\rho L\mathcal{H}_0^2/(\eta_0\mathcal{S}_0^2)]$', fontsize=22)
plt.xticks((1e-2, 1e-1, 1e0, 1e1, 1e2), size=13)
ax.set_ylim(0.3, 1.05)
plt.ylabel(r'$\mathcal{H}/\mathcal{H}_0$', fontsize=22)
plt.yticks((0.4, 0.6, 0.8, 1.0), size=13)
plt.legend(loc='lower left', fontsize=15)
plt.show()
```

The figure below plots porosity versus time for three values of $\cmpdisspar$.

```python
f, ax = plt.subplots()
zoom = 2.0
f.set_size_inches(5.4 * zoom, 3.1 * zoom)
f.set_facecolor('w')

y0 = [1., 0.0]  # initial condition: H=1, phi=0.0
lambda_ = 27.0
Pi = np.asarray([0.1, 1., 10.])
tmax = np.asarray([60., 10., 2.6])
linesty = ['-', '--', '-.']
for pi, tm, ls in zip(Pi, tmax, linesty):
    t = np.arange(0.0, tm, 0.05)
    S = odeint(thick, y0, t, args=(pi, lam), rtol=1e-8)
    plt.loglog(t, S[:, 1], 'k', linestyle=ls, linewidth=2, label=f'$\Pi={str(pi)}$')

plt.xlabel(r'$t/[\rho L\mathcal{H}_0^2/(\eta_0\mathcal{S}_0^2)]$', fontsize=22)
plt.xticks((1e-2, 1e-1, 1e0, 1e1, 1e2), size=13)
plt.ylabel(r'$\phi$', fontsize=22)
plt.yticks((1e-2, 1e-1), size=13)
plt.legend(loc='lower right', fontsize=15)
plt.legend()
plt.show()
```
