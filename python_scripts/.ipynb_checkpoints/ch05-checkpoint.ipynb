## The dihedral angle 

Interfaces oriented according to angles $\Theta_k$ and interfacial tensions $\sigma_k$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

f, ax = plt.subplots()
f.set_facecolor('w')
zoom = 3.0
f.set_size_inches(8.3 * zoom, 3.8 * zoom)
f.set_facecolor('w')

theta_k = 30.

theta = theta_k * np.pi / 180.
ct = np.cos(theta)
st = np.sin(theta)
x = [-ct, 0.0, ct, 0.0, ct]
y = [0.0, 0.0, st, 0.0, -st]
plt.plot(x, y, '-k', linewidth=2)
plt.plot(x, y, 'bo')
ax.set_xlim(-1.0, 1.0)
ax.set_ylim(-1.0, 1.0)

plt.annotate(r'$\Theta_1$', xy=[0.0, 0.05], 
             horizontalalignment='right', verticalalignment='bottom', 
             fontsize=28)
plt.annotate(r'$\Theta_2$', xy=[0.0, -0.05], 
             horizontalalignment='right', verticalalignment='top', 
             fontsize=28)
plt.annotate(r'$\Theta_3$', xy=[0.1, -0.01], 
             horizontalalignment='left', verticalalignment='center', 
             fontsize=28)

f = 0.9
plt.annotate(r'$\sigma_1$', xy=(f*ct, -f*st-0.18), fontsize=28, rotation=-theta_k/2.0)
plt.annotate(r'$\Longrightarrow$', xy=(f*ct, -f*st-0.13), fontsize=28, rotation=-theta_k/2.0)

plt.annotate(r'$\sigma_2$', xy=(f*ct, f*st+0.12), fontsize=28, rotation=theta_k/2.0)
plt.annotate(r'$\Longrightarrow$', xy=(f*ct, f*st+0.05), fontsize=28, rotation=theta_k/2.0)

plt.annotate(r'$\sigma_3$', xy=(-ct+0.02, 0.07), fontsize=28, rotation=0.0)
plt.annotate(r'$\Longleftarrow$', xy=(-ct, 0.0), fontsize=28, rotation=0.0)

ax.set_axis_off()
```

An equation for dihedral angle $\dihed$ is
\begin{equation}
  \label{eq:dihedral-angle}
  \cos\frac{\dihed}{2} = \frac{\sfce\solsol}{2\sfce\liqsol}.
\end{equation}



It follows below the outline of a two-dimensional pore at a triple junction. This can be considered to be a slice through the three-dimensional pore shown above, taken at the pore throat, half way along the triple junction. The dihedral angle is $\Theta=30^\circ$. Interfaces are labelled as liquid-solid (*ls*) and solid-solid (*ss*).

```python
theta = 30.0 * np.pi / 180.  # the dihedral angle
```

```python
f, ax = plt.subplots()
zoom = 3.0
f.set_size_inches(8.3 * zoom, 3.8 * zoom)
f.set_facecolor('w')

Theta = 120.*np.pi/180./2.
ysh = 0.2

xI = 0.3 
x = np.linspace(0.0, xI, 100)
y = 0.5 * xI * ((x**2/xI**2 - 1.)*np.tan(np.pi/6. - 0.5*theta) + 2.0*np.tan(np.pi/6.))
X = np.zeros(x.shape[0] * 2 * 2).reshape(2, 2*x.shape[0])
X[0, :] = np.concatenate((-x[::-1], x), axis=0)
X[1, :] = np.concatenate((y[::-1], y), axis=0)
R = np.array(((np.cos(Theta*2.), -np.sin(Theta*2)), (np.sin(Theta*2), np.cos(Theta*2))))
RX = np.fliplr(np.dot(R, X))
RRX = np.fliplr(np.dot(np.linalg.inv(R), X))
x = np.concatenate((np.fliplr(RX)[0, :], x, np.fliplr(RRX)[0, :]), axis=0)
y = np.concatenate((np.fliplr(RX)[1, :], y, np.fliplr(RRX)[1, :]), axis=0)
plt.plot(x, y+ysh, '-k', linewidth=2)

cT = np.cos(Theta) 
sT = np.sin(Theta)
ss = np.array(((0.0, 0.0), (-1.0, np.amin(y))))
plt.plot(ss[0, :], ss[1, :]+ysh, '-k', linewidth=2)
ss = np.dot(np.linalg.inv(R), ss)
plt.plot(ss[0, :], ss[1, :]+ysh, '-k', linewidth=2)
ss = np.dot(np.linalg.inv(R), ss)
plt.plot(ss[0, :], ss[1, :]+ysh, '-k', linewidth=2)

plt.plot([0, 0.5], [ysh, ysh], '-k', linewidth=0.5)
plt.plot([0, 0], [ysh, 0.5+ysh], '-k', linewidth=0.5)
plt.annotate(r'$x$', xy=(0.65, ysh), verticalalignment='top', horizontalalignment='left', fontsize=15)
plt.annotate(r'$y$', xy=(0.0, 0.65+ysh), verticalalignment='bottom', horizontalalignment='left', fontsize=15)

f = 0.25
vx = [-f*np.sin(theta/2), 0.0, f*np.sin(theta/2)]
vy = [f*np.cos(theta/2) + np.amin(y) +ysh, np.amin(y) + ysh, f*np.cos(theta/2) + np.amin(y) + ysh]
plt.plot(vx, vy, '-k', linewidth=0.1)
plt.annotate(r'$\Theta$', xy=[0.001, np.amin(y)+ysh+0.09], verticalalignment='bottom', 
             horizontalalignment='center', fontsize=15)

plt.annotate(r'$\ell s$', xy=(-0.1, ysh), verticalalignment='center', horizontalalignment='right', 
             fontsize=15, rotation=-55.)
plt.annotate(r'$ss$', xy=(0.5, 0.33+ysh), verticalalignment='center', horizontalalignment='right', 
             fontsize=15, rotation=30.)
plt.annotate(r'$ss$', xy=(-0.5, 0.33+ysh), verticalalignment='center', horizontalalignment='left', 
             fontsize=15, rotation=-30.)
plt.annotate(r'$ss$', xy=(-0.01, -0.5+ysh), verticalalignment='bottom', horizontalalignment='right', 
             fontsize=15, rotation=90.)

ax.set_xlim(-1.0, 1.0)
ax.set_ylim(-1.0, 1.0)
ax.set_axis_off()

plt.show()
```

## Permeability

A simple model which represents the permeability at some reference value of porosity $\phi_0$ is

\begin{equation}
  \label{eq:permeability-simple}
  \perm = k_0(\por/\por_0)^\permexp.
\end{equation}

Equation \eqref{eq:permeability-simple} is plotted below for two values of n. Also shown are permeabilities computed on the basis of simulation of ow through pores of a measured pore geometry.

```python
import csv

f, ax = plt.subplots()
zoom = 3.0
f.set_size_inches(5.2 * zoom, 4.9 * zoom)
f.set_facecolor('w')

phi = np.logspace(-3.0, 0.0, 1000, endpoint=False)

phid = 0.25
phi0 = 0.01
K0 = 1e-12

f = phi/phi0

K2 = K0 * (f ** 2.7)
K3 = K0 * (f ** 3.0)
Kkc = K0 * (f ** 3.0) / ((1. - phi)/(1. - phi0)) ** 2

plt.loglog(phi, K2, '-k', label=r'$k_0(\phi/\phi_0)^{2.7}$', linewidth=2)
plt.loglog(phi, K3, '--k', label=r'$k_0(\phi/\phi_0)^3$', linewidth=2)
plt.loglog(phi, Kkc, '-.k', label=r'$k_0(\phi/\phi_0)^3[(1-\phi)/(1-\phi_0)]^{-2}$', linewidth=2)

# data and data fit
grainsize = 35e-6
C = 58
permexp = 2.7
dmphi = np.asarray([0.01, 0.3])
dmk = grainsize ** 2 / C * (dmphi ** 2.7)
plt.loglog(dmphi, dmk, '-', color=[0.5, 0.5, 0.5], linewidth=2, label=r'$d^2\phi^{2.7}/C$')

with open('Data_Permeability_Miller2015.csv', mode='r') as csv_file:
    data = csv.reader(csv_file)
    lst = [row for row in data]
data = np.asarray([[float(l) for l in l_] for l_ in lst[1:]])

plt.loglog(data[:, 0], data[:, -1], 'ok', label=r'Miller et al. 2015')
for d in data:
    plt.plot([d[0]+d[1], d[0]+d[2]], [d[3], d[3]], '-k', linewidth=1)

plt.plot([phid, phid], [1e-16, 1.0], ':k', label=r'$\phi_\Xi$')

ax.set_xlim(1e-3, 1.0)
ax.set_ylim(1e-16, 1.0)
plt.xlabel(r'$\phi$', fontsize=20)
plt.ylabel(r'$k_\phi$, $m^2$', fontsize=20)
plt.legend(loc='upper left', fontsize=15)

plt.show()
```
